A common initial piece of exploratory analysis is the histogram. With n=5000 integers,

````python
from prisoner_puzzle import monte_carlo
import matplotlib.pyplot as plt 

results = monte_carlo(5000,50)
plt.hist(results, bins=30)
plt.title("Prisoner Escape Riddle Historgram")
plt.xlabel("Number of days to escape")
plt.ylabel("Frequency")
plt.show()
````

There at least two things to note about this histogram. First, the range appears to be from 
about 1600 to about 4100. Second, the histogram is not exactly centered in this range, 
but skews to the left. Put another way, the tail is longer on the right than on the left. 
This means that across the range of 5,000 experiments, the average number of days for prisoner 
release was less than the median number of days. Let's add some code to look at the actual min, 
max, mean, and median, and put it all into a human-readable format:

````python
from prisoner_puzzle import monte_carlo
import numpy as np
import matplotlib.pyplot as plt 

number_of_experiments = 5000
number_of_prisoners = 50
results = monte_carlo(number_of_experiments, number_of_prisoners)
#Calculate some statistics:
minimum = min(results)
maximum = max(results)
median = (maximum - minimum)/2.0 + minimum
mean = np.mean(results)

string = "The simulation with {} results took an average of {} days  \
  or {} months or {} years to finish."
print(string.format(number_of_experiments, mean, mean/12, mean/365.0))

#what's the spread?
print "max:", maximum
print "min:", minimum

#where does the mean fall within that spread?
print "mean:", mean
print "median:", median
print "distance from mean to median:", mean - median
#Plot a histogram
histogram_data = plt.hist(results, bins=30)
plt.vlines(
  median, 
  0,  
  max(histogram_data[0]), 
  linestyles='dashed', 
  lw = 4,  
  label = "median", 
  color='red')
plt.vlines(
  mean, 
  0,  
  max(histogram_data[0]),
  linestyles='dashed',
  lw = 4,
  label = "mean",
  color='orange')
plt.title("Prisoner Escape Riddle Historgram")
plt.xlabel("Number of days to escape")
plt.ylabel("Frequency")
plt.legend()
plt.show()

""""
And this code produces the following:
The simulation with 50 results took an average of 2736.36 days or 228.03 months or 7.49687671233 years to finish.
max: 3598
min: 1967
mean: 2736.36
median: 2782.5
distance from mean to median: -46.14
"""
````

If we use a huge number of samples--say 20 million--then we'd have a truly representative average value.
Taking one more sample would have virtually no effect on the calculated average. 
If we took only three or four samples, then we couldn't be too sure that our random number generator 
didn't give us lucky draws and allow the prisoners to escape quickly. So how close to 20 million

````python
# Iteration 4: Monte Carlo simulator to method; parameterize no. of prisoners 
# and no. of runs; prettify output

from numpy import array, mean
import random

def run_simulation(number_of_prisoners):
  counter = 0 
  day_number = 0 
  prisoner_chief = 1 
  lever_position = 'down'
  prisoners_who_have_operated_lever = []

  while counter < number_of_prisoners:
    prisoner_number = random.randint(1,number_of_prisoners) # Select a prisoner
    if not prisoner_number in prisoners_who_have_operated_lever:
      if lever_position == 'down':
        prisoners_who_have_operated_lever.append(prisoner_number)
        lever_position = 'up'
    if prisoner_number == prisoner_chief:
      if lever_position == 'up':
          lever_position = 'down'
          counter += 1
    day_number += 1
  return day_number 

def monte_carlo(number_of_experiments, prisoner_count):
  simulation_results = []
  simulations = 0 
  while simulations < number_of_experiments:
    simulation_results.append(run_simulation(prisoner_count))
    simulations += 1
  simulation_results = array(simulation_results) #convert to numpy array
  return simulation_results

monte_500 = monte_carlo(500,50)
average = mean(monte_500)
text = "The simulation with {} results took an average of {} days or {} months or {} years to finish."
print(text.format(500, average, average/12, average/365.0))
Convergence
````
